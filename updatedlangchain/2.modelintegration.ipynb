### Streaming and Batch

#### Streaming
###### Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particulary for longer responses. Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time 

```python
model.invoke("write for me a 200 words paragraph on agentic AI")
```

```python
model.stream("write for me a 200 words paragraph on agentic AI")
```

```python
for chunk in model.stream("write for me a 200 words paragraph on agentic AI"):
    print(chunk.text, end="|",flush=True)
```

```python
model.invoke("why do parrots have colorful feathers?")
```

```python
for chunk in model.stream("why do parrots have colorful feathers?"):
    print(chunk.text, end="|",flush=True)
```

#### Batch
##### Batching is a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel  

```python
responses = model.batch(["Whay do parrots have colorful feathers?", 
                         "What is quantum computing?", 
                         "What is the meaning of life?"])
for response in responses:
    print(response.content)
```

```python
responses = model.batch(["Whay do parrots have colorful feathers?", 
                         "What is quantum computing?", 
                         "What is the meaning of life?"],
                         config={"max_concurrency": 5})
for response in responses:
    print(response.content)
```
